```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(ndf[['D','A','S','IA']])
das=scaler.transform(ndf[['D','A','S','IA']])
dast=das.transpose()
ndf['d1']=dast[0]
ndf['a1']=dast[1]
ndf['s1']=dast[2]
ndf['IA1']=dast[3]
```

```python
ndf['Urban/Rural']=''
for i in range(len(ndf)):
    if ndf['UR'].iloc[i]==0:
        ndf['Urban/Rural'].iloc[i]='Urban'
    else:
        ndf['Urban/Rural'].iloc[i]='Rural'

        ndf['ias']=''
for i in range(len(ndf)):
    if ndf['IA sev2'].iloc[i]==0:
        ndf['ias'].iloc[i]='Normal'
    else:
        ndf['ias'].iloc[i]='Addicted'
        
ndf['IAD']=ndf['ias']
ndf['D sev2']=ndf['D sev']
ndf['A sev2']=ndf['A sev']
ndf['S sev2']=ndf['S sev']

for j in ['D sev2', 'A sev2', 'S sev2']:
    for i in range(len(ndf[j])):
        if ndf[j].iloc[i] == 1.0:
            ndf[j].iloc[i] = 'Normal'
        if ndf[j].iloc[i] == 2.0:
            ndf[j].iloc[i] = 'Mild'
        if ndf[j].iloc[i] == 3.0:
            ndf[j].iloc[i] = 'Moderate'
        if ndf[j].iloc[i] == 4.0:
            ndf[j].iloc[i] = 'Severe'
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import sklearn



xu=ndf[ndf['UR']==0]
xr=ndf[ndf['UR']==1]

X = x = xr[['d1', 'a1', 's1',
       'MI2', 'Q11',  'Q16',
        'Q18',   'Q24',  'AI','Q9','Q2','Q13', 'Q21','Q4', 'Q6', 'Q7']]
#X = x = xu[['d1', 'a1', 's1','Q18','Q16','MI2','Q11','AI','Q13']]
#X=x=xr[['d1','s1','a1','Q21','Q16','Q24','AI','Q2','Q9']]
for i in ['d1','s1','a1','Q16']:
    X[i]=X[i].astype('float64')

y = xr["IA sev2"]
#y=(y=='Addicted').astype('float32')
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
smote_enn = SMOTEENN(random_state=101)
oversample = SMOTE(random_state=101)
#X_r, y_r = oversample.fit_resample(X, y)
#X_resampled, y_resampled = smote_enn.fit_resample(X, y)

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)

X_r, y_r = oversample.fit_resample(x_train, y_train)

#trainX,testX, trainy,  testy = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import numpy as np

rf = RandomForestClassifier(n_estimators=1000,  n_jobs=-1, random_state=101)
rf.fit(X_r, y_r)
y_pred=rf.predict(x_test)

print(metrics.f1_score(y_test, y_pred, average='macro'), metrics.accuracy_score(y_test, y_pred))

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
#print(confusion_matrix(y_test, y_pred))

import matplotlib.pyplot as plt
std = np.std([tree.feature_importances_ for tree in rf.estimators_],
             axis=0)
hh = rf.feature_importances_

indices = np.argsort(hh)[::-1]

dn={
    'Q2':'2',
    'd1':'Depression',
    's1':'Stress',
    'a1':'Anxiety',
    'Q9':'11',
    'Q11':'10',
    'Q13':'8',
    'Q16':'7',
    'Q18':'9',
    'Q21':'12',
    'Q24':'13',
    'AI':'1',
    'MI2':"3",
    'Q4':'4',
    'Q6': '5',
    'Q7':'6'
}

# Print the feature ranking
print("Feature ranking:")
tks=[]
for f in range(x.shape[1]):
    tks.append(dn[x.columns[indices[f]]])
    print("%d. feature %s (%f)" % (f + 1, x.columns[indices[f]], hh[indices[f]]))

# Plot the impurity-based feature importances of the forest
plt.figure()
plt.title("Feature importance for rural data",fontsize=18)
# plt.bar(['Stress', 'Depression','Anxiety', 'Over-use of technology', 'Calls per day', 'Frequency of notifications', 'Monthly Income',  'Internet expenses', 'Age'], hh[indices],
#         color="r", yerr=std[indices], align="center")

hh.sort()
tks.reverse()

plt.barh( tks,hh, 
         color="b",  align="center")
plt.xticks(rotation=90)
for i in range(0,len(x.columns)):
    print(i,x.columns[i])
plt.ylim([-1, X.shape[1]])
plt.ylabel('Features',fontsize=18)
plt.xlabel('Importance',fontsize=18)

plt.savefig('rural v_imp.jpg', dpi=500,bbox_inches = 'tight')
plt.show()
```

```python
from tensorflow import keras
from sklearn.datasets import make_classification
from sklearn.metrics import roc_auc_score
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LeakyReLU

xu=ndf[ndf['UR']==0]
xr=ndf[ndf['UR']==1]

X = x = xu[['d1','s1','a1','Q18','Q16','MI2','Q11','AI','Q13']]
#X = x = xr[['d1','s1','a1','Q21','Q16','Q24','AI','Q2','Q9']]
#for i in ['AI','Q2(Gender 1 male)','D sev','A sev' ,'S sev']:
for i in ['d1','s1','a1','Q16','AI']:
    X[i]=X[i].astype('float64')

y = xu["IA sev2"]
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
smote_enn = SMOTEENN(random_state=101)
oversample = SMOTE(random_state=101)
#X_r, y_r = oversample.fit_resample(X, y)
#X_resampled, y_resampled = smote_enn.fit_resample(X, y)

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)

X_r, y_r = oversample.fit_resample(x_train, y_train)


from sklearn.datasets import make_classification
from sklearn.metrics import roc_auc_score
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential

opt = keras.optimizers.Adam(learning_rate=0.001)

# define the neural network model
n_layers=range(2000,10,5)
n_input = X_r.shape[1]

# define model
model = Sequential()
# define first hidden layer and visible layer
model.add(Dense(9, input_dim=n_input, kernel_initializer='he_uniform'))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(120, kernel_initializer='he_uniform',kernel_regularizer='l1_l2'))
model.add(LeakyReLU(alpha=0.1))
model.add(Dropout(0.3))
for i in n_layers:        
    model.add(Dense(i,  kernel_initializer='he_uniform'))
    model.add(LeakyReLU(alpha=0.1))
    model.add(Dropout(0.4))

model.add(Dense(100, kernel_initializer='he_uniform'))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(100, kernel_initializer='he_uniform'))
model.add(LeakyReLU(alpha=0.1))
model.add(Dropout(0.5))
model.add(Dense(6, kernel_initializer='he_uniform', kernel_regularizer='l1_l2'))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(1, activation='sigmoid'))
# define loss and optimizer
model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
    
    

# prepare dataset
# define the model
#model = define_model(n_input, np.linspace(500,20,50))
# fit model
history = model.fit(X_r, y_r, epochs=2000)
# make predictions on the test dataset
yhat = model.predict(x_test)
# evaluate the ROC AUC of the predictions
score = roc_auc_score(y_test, yhat)
print('ROC AUC: %.3f' % score)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import sklearn

xu=ndf[ndf['UR']==0]
xr=ndf[ndf['UR']==1]
# X=xu[['Q2', 'Q7', 'd1', 'a1', 's1',
#        'Q9', 'Q11', 'Q13', 'Q16',
#         'Q18', 'Q21',  'Q24',  'AI',
#        'MI2']]
X = x = xr[['d1', 'a1', 's1',
       'MI2', 'Q11',  'Q16',
        'Q18',   'Q24',  'AI','Q9','Q2','Q13', 'Q21','Q4', 'Q6', 'Q7']]
#['Gender', 'AI', 'Depression','Stress','Calls per day', 'Anxiety','Do you feel anxious without your phone','Type of mobile phone','Over-use of technology']
#X = x = xr[['Q2', 'AI','d1','s1','Q16','a1', 'Q21', 'Q9', 'Q18']]
#X = x = xu[['d1', 'a1', 's1','Q18','Q16','MI2','Q11','AI','Q13']]
#X=x=xr[['d1','s1','a1','Q21','Q16','Q24','AI','Q2','Q9']]
for i in ['d1','s1','a1','Q16']:
    X[i]=X[i].astype('float64')

y = xr["IA sev2"]
#y=(y=='Addicted').astype('float32')
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
smote_enn = SMOTEENN(random_state=101)
oversample = SMOTE(random_state=101)
#X_r, y_r = oversample.fit_resample(X, y)
#X_resampled, y_resampled = smote_enn.fit_resample(X, y)

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)

X_r, y_r = oversample.fit_resample(x_train, y_train)

#trainX,testX, trainy,  testy = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import numpy as np
rf = RandomForestClassifier(n_estimators=1000,  n_jobs=-1, random_state=101)
rf.fit(X_r, y_r)
y_pred=rf.predict(x_test)

print(metrics.f1_score(y_test, y_pred, average='macro'), metrics.accuracy_score(y_test, y_pred))
```
